```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
from sklearn.compose import ColumnTransformer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
dataset = pd.read_csv('life_expectancy.csv')
dataset.head()
```

```python
dataset.describe()
```

```python
dataset = dataset.drop(['Country'], axis=1)
```

## Select features and labels

```python
features = dataset.iloc[:, :-2]
labels = dataset.iloc[:, -1]
```

```python
features
```

## Encode categorical columns

```python
features = pd.get_dummies(features)
features = features.iloc[:, :-1]

features.head()
```

```python
# change status developed col dtype to int64
features.Status_Developed = features.Status_Developed.astype('int64')
```

```python
features.dtypes
```

## Split data into train-test

```python
features_train, features_test, labels_train, labels_test = train_test_split(features, labels, test_size = 0.20, random_state=0)
```

## Standardize Data

```python
numerical_features = features.select_dtypes(include=['float64', 'int64'])
numerical_columns = numerical_features.columns
ct = ColumnTransformer([("only numeric", StandardScaler(), numerical_columns)], remainder='passthrough')
```

```python
features_train_scaled = ct.fit_transform(features_train)
features_test_scaled = ct.transform(features_test)
```

## Implementing Deep Learning

```python
# initialize model obj
my_model = Sequential()

# create input layer with shape of number of features
input = InputLayer(input_shape = (features.shape[1], ))
my_model.add(input)

# add hidden layer with 64 units
my_model.add(Dense(64, activation = "relu"))

# add one output layer with 1 output
my_model.add(Dense(1))

# summary of model
my_model.summary()
```

## Initializing the optimizer and compiling the model

```python
# Create an instance of the Adam optimizer with the learning rate equal to 0.01
opt = Adam(learning_rate = 0.01)

#  compile the model
my_model.compile(loss = 'mse', metrics = ['mae'], optimizer = opt)

# Fit and evaluate the model
my_model.fit(features_train_scaled, labels_train, epochs = 40, batch_size = 1, verbose = 1)
```

```python
# testing model
res_mse , res_mae = my_model.evaluate(features_test_scaled, labels_test, verbose = 0)
print(res_mse, res_mae)
```
